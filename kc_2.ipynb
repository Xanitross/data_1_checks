```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as pt
```

```python
df = pd.read_excel(".\Assets\sqf-2021.xlsx")
df.set_index("STOP_ID", inplace=True)
df
```

Bar Graph showing stops by day of week

```python
df["DAY2"].value_counts().plot(kind="bar", title="Number of Stops by Day of the Week")
```

Exact values:

```python
print(df.value_counts("DAY2"))
```

Cleaning data to remove the year from every date entry:

```python
df["STOP_FRISK_DATE"] = df["STOP_FRISK_DATE"].str.slice(5)
```

Graph showing stops for the year:

```python
print(df.value_counts("STOP_FRISK_DATE"))
```

```python
df["STOP_FRISK_DATE"].value_counts()[df["STOP_FRISK_DATE"].unique()].plot(kind="line", title="Stops for 2021", figsize=(10,5))
```
